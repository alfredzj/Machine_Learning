```python
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pathlib import Path

# preprocess
from sklearn.preprocessing import StandardScaler, LabelEncoder

# metrics
from sklearn.metrics.pairwise import pairwise_distances
from sklearn.metrics import silhouette_score, adjusted_rand_score, f1_score

# clustering
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture

# dimensionality reduction
from sklearn.decomposition import PCA, FastICA
from sklearn.random_projection import GaussianRandomProjection
from sklearn.ensemble import RandomForestClassifier

# nn related
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_validate

# Ignore  the warnings
import warnings
warnings.filterwarnings('always')
warnings.filterwarnings('ignore')
```

#### Load the data

```python
# read data
df_pw = pd.read_csv(Path("phishing_website.csv"))

# get features
features_pw = list(df_pw.columns[:-1])

# get X and y
Xpw = df_pw.drop('Result', axis=1).values
ypw = df_pw['Result'].values

# preprocess X
sc = StandardScaler()
Xpw = sc.fit_transform(Xpw)
```

```python
# read data
df_rw = pd.read_csv(Path("winequality-red.csv"))

# remove unnecessary cols based previous experiment
df_rw.drop(["residual sugar",'free sulfur dioxide','pH'],axis = 1,inplace = True)

# bin "quality" variable into three levels: bad/good
bins = [0, 6.5, 10]
labels = ["bad","good"]
df_rw['binned_quality'] = pd.cut(df_rw['quality'], bins=bins, labels=labels)
df_rw.drop('quality',axis =1, inplace = True)

# encode the quality col
label_quality = LabelEncoder()
df_rw['binned_quality'] = label_quality.fit_transform(df_rw['binned_quality'])

# get features
features_rw = list(df_rw.columns[:-1])

# get X and y
Xrw = df_rw.drop( "binned_quality",axis=1).values
yrw = df_rw["binned_quality"].values

# preprocess X
sc = StandardScaler()
Xrw = sc.fit_transform( Xrw )
```

#### Run the clustering algo: K-means, EM

```python
def plot_clusters(**kargs):
    plt.figure()
    for data_label, data_tuple in kargs.items():
        xvec, yvec = data_tuple
        plt.plot(xvec, yvec, label=data_label)
    plt.xlabel('n_cluster')
    plt.ylabel('metrics')
    plt.legend(loc='best')
    plt.show()
    
def run_kmeans(X, y, plot=True):
    
    np.random.seed(0)
    sil_metrics, rand_metrics, t_metrics = [], [], []
    num_k = list(range(2, 30))

    for k in num_k:
        start_time = time.time()
        km = KMeans(n_clusters=k, n_init=10,random_state=100, n_jobs=-1).fit(X)
        end_time = time.time() 
        sil_metrics.append(silhouette_score(X, km.labels_))  
        rand_metrics.append(adjusted_rand_score(y, km.labels_))
        t_metrics.append(end_time - start_time)
        
    if plot:     
        score_info = {'silhouette_score': [num_k, sil_metrics], 
                       'adjusted_rand_score': [num_k, rand_metrics]}
        time_info = {'computation_time': [num_k, t_metrics]}
        plot_clusters(**score_info)
        plot_clusters(**time_info)
        
    return num_k, sil_metrics, rand_metrics, t_metrics

def run_EM(X, y, plot=True):

    np.random.seed(0)
    sil_metrics, rand_metrics, t_metrics = [], [], []
    num_c = list(range(2, 30))
    
    for c in num_c:
        start_time = time.time()
        em = GaussianMixture(n_components=c, covariance_type='diag', warm_start=True, random_state=100).fit(X)
        end_time = time.time() 
        
        labels = em.predict(X)
        sil_metrics.append(silhouette_score(X, em.predict(X)))  
        rand_metrics.append(adjusted_rand_score(y, em.predict(X)))
        t_metrics.append(end_time - start_time)
        
    if plot:     
        score_info = {'silhouette_score': [num_c, sil_metrics], 
                       'adjusted_rand_score': [num_c, rand_metrics]}
        time_info = {'computation_time': [num_c, t_metrics]}
        plot_clusters(**score_info)
        plot_clusters(**time_info)
        
    return num_c, sil_metrics, rand_metrics, t_metrics
```

```python
num_c_kpw, sil_metrics_kpw, rand_metrics_kpw, t_metrics_kpw = run_kmeans(Xpw,ypw)
```

```python
num_c_krw, sil_metrics_krw, rand_metrics_krw, t_metrics_krw = run_kmeans(Xrw,yrw)
```

#### Apply the dimensionality reduction algo

```python
def run_PCA(X, y):
    
    pca = PCA(random_state=100).fit(X)
    cum_var = np.cumsum(pca.explained_variance_ratio_)

    fig, ax1 = plt.subplots()
    ax1.plot(list(range(len(pca.explained_variance_ratio_))), cum_var, 'b-')
    ax1.set_xlabel('num_principal_components')
    ax1.set_ylabel('cum_explained_variance_ratio', color='b')
    ax1.tick_params('y', colors='b')

    ax2 = ax1.twinx()
    ax2.plot(list(range(len(pca.singular_values_))), pca.singular_values_, 'r-')
    ax2.set_ylabel('Eigenvalues', color='r')
    ax2.tick_params('y', colors='r')

    fig.tight_layout()
    plt.show()
```

```python
run_PCA(Xpw, ypw)
```

```python
run_PCA(Xrw, yrw)
```

```python
def run_ICA(X, y):
    
    num_ic = list(np.arange(2,(X.shape[1]+1),2))
    ica = FastICA(random_state=100)
    kurt_metrics = []

    for ic in num_ic:
        ica.set_params(n_components=ic)
        transformed_x = pd.DataFrame(ica.fit_transform(X))
        xkurt = transformed_x.kurt(axis=0)
        kurt_metrics.append(xkurt.abs().mean())

    plt.figure()
    plt.xlabel('independent_components')
    plt.ylabel('avg_kurtosis')
    plt.plot(num_ic, kurt_metrics)
    plt.show()
```

```python
run_ICA(Xpw, ypw)
```

```python
run_ICA(Xrw, yrw)
```

```python
def pairwise_distcorr(X1,X2):    
    d1 = pairwise_distances(X1).ravel()
    d2 = pairwise_distances(X2).ravel()
    return np.corrcoef(d1,d2)[0,1]

def run_RP(X, y):
    
    num_c = list(np.arange(2,(X.shape[1]+1),2))
    distcorr_metrics = []
    
    for c in num_c:
        rp = GaussianRandomProjection(random_state=100, n_components=c)
        distcorr_metrics.append(pairwise_distcorr(rp.fit_transform(X), X))

    plt.figure()
    plt.xlabel('n_components')
    plt.ylabel('dist_corr')
    plt.plot(num_c, distcorr_metrics)
    plt.show()
```

```python
run_RP(Xpw, ypw)
```

```python
run_RP(Xrw, yrw)
```

```python
def run_RFC(X, y, features, threshold=0.95):
    rfc = RandomForestClassifier(n_estimators=500,min_samples_leaf=round(len(X)*.01),random_state=100,n_jobs=-1)
    f_impt = rfc.fit(X,y).feature_importances_ 
    df_impt = pd.DataFrame(f_impt,columns=['feature_importance'],index=features)
    df_impt.sort_values(by=['feature_importance'],inplace=True,ascending=False)
    df_impt['cum_sum'] = df_impt['feature_importance'].cumsum()
    df_impt = df_impt[df_impt['cum_sum']<=threshold]
    return df_impt
```

#### Rerun neural network based on newly projected data

```python
def plot_nncurve(x, y1, y1_std, y2, y2_std):
    
    x = np.array(x)
    y1, y1_std = np.array(y1), np.array(y1_std)
    y2, y2_std = np.array(y2), np.array(y2_std)
    
    plt.figure()
    plt.xlabel('training_size')
    plt.ylabel('f1_score')
    plt.fill_between(x, y1 - 2*y1_std, y1 + 2*y1_std, alpha=0.1, color="b")
    plt.fill_between(x, y2 - 2*y2_std, y2 + 2*y2_std, alpha=0.1, color="r")
    plt.plot(x, y1, 'o-', color="b", label="train_score")
    plt.plot(x, y2, 'o-', color="r", label="test_score")
    plt.legend(loc="best")
    plt.show()
      
def plot_nntime(x, y):  
    plt.figure()
    plt.xlabel('training_size')
    plt.ylabel('fit_time')
    plt.plot(x, y)
    plt.show()

def run_nn(clf, X, y, plot=True):
    
    np.random.seed(0)
    train_score, train_score_std = [], []
    test_score, test_score_std = [], []
    fit_time = []
    train_sizes=(np.linspace(.05, 1.0, 20)*len(y)).astype('int')  
    
    for i in train_sizes:
        idx = np.random.randint(X.shape[0], size=i)
        X_subset = X[idx,:]
        y_subset = y[idx]
        scores = cross_validate(clf, X_subset, y_subset, cv=5, scoring='f1', n_jobs=-1)
        
        train_score.append(np.mean(scores['train_score']))
        train_score_std.append(np.std(scores['train_score']))
        test_score.append(np.mean(scores['test_score']))
        test_score_std.append(np.std(scores['test_score']))
        fit_time.append(np.mean(scores['fit_time']))
        
    if plot:
        plot_nncurve(train_sizes, train_score, train_score_std, test_score, test_score_std)
        plot_nntime(train_sizes, fit_time)
    
    return train_sizes, train_score, test_score, fit_time
```

```python
learner_pw = MLPClassifier(hidden_layer_sizes=(50,), solver='adam', activation='relu', learning_rate_init=0.05, random_state=100)
trs_pw, trsc_pw, tsc_pw, time_pw = run_nn(learner_pw, Xpw, ypw)
trs_pw_pca, trsc_pw_pca, tsc_pw_pca, time_pw_pca = run_nn(learner_pw, pca_pw, ypw)
trs_pw_ica, trsc_pw_ica, tsc_pw_ica, time_pw_ica = run_nn(learner_pw, ica_pw, ypw)
trs_pw_rp, trsc_pw_rp, tsc_pw_rp, time_pw_rp = run_nn(learner_pw, rp_pw, ypw)
trs_pw_rfc, trsc_pw_rfc, tsc_pw_rfc, time_pw_rfc = run_nn(learner_pw, rfc_pw, ypw)
```

```python
learner_rw = MLPClassifier(hidden_layer_sizes=(11,), solver='adam', activation='relu', learning_rate_init=0.05, random_state=100)
trs_rw, trsc_rw, tsc_rw, time_rw = run_nn(learner_rw, Xrw, yrw)
trs_rw_pca, trsc_rw_pca, tsc_rw_pca, time_rw_pca = run_nn(learner_rw, pca_rw, yrw)
trs_rw_ica, trsc_rw_ica, tsc_rw_ica, time_rw_ica = run_nn(learner_rw, ica_rw, yrw)
trs_rw_rp, trsc_rw_rp, tsc_rw_rp, time_rw_rp = run_nn(learner_rw, rp_rw, yrw)
trs_rw_rfc, trsc_rw_rfc, tsc_rw_rfc, time_rw_rfc = run_nn(learner_rw, rfc_rw, yrw)
```

```python
def compare_nnscore(n, original, pca, ica, rp, rfc, ylabel):   
    plt.figure()
    plt.xlabel('training_size')
    plt.ylabel(ylabel)
    plt.plot(n, original, '-', color="black", label="Original")
    plt.plot(n, pca, '-', color="b", label="PCA")
    plt.plot(n, ica, '-', color="r", label="ICA")
    plt.plot(n, rp, '-', color="g", label="RP")
    plt.plot(n, rfc, '-', color="m", label="RFC")
    plt.legend(loc="lower right")
    plt.show()
```

```python
compare_nnscore(trs_pw, trsc_pw, trsc_pw_pca, trsc_pw_ica, trsc_pw_rp, trsc_pw_rfc, 'train_f1_score')
compare_nnscore(trs_pw, tsc_pw, tsc_pw_pca, tsc_pw_ica, tsc_pw_rp, tsc_pw_rfc, 'test_f1_score')
compare_nnscore(trs_pw, time_pw, time_pw_pca, time_pw_ica, time_pw_rp, time_pw_rfc, 'fit_time')
```

```python
compare_nnscore(trs_rw, trsc_rw, trsc_rw_pca, trsc_rw_ica, trsc_rw_rp, trsc_rw_rfc, 'train_f1_score')
compare_nnscore(trs_rw, tsc_rw, tsc_rw_pca, tsc_rw_ica, tsc_rw_rp, tsc_rw_rfc, 'test_f1_score')
compare_nnscore(trs_rw, time_rw, time_rw_pca, time_rw_ica, time_rw_rp, time_rw_rfc, 'fit_time')
```

#### Clustering on DR ->NN


```python
def add_clusters(X, cluster_method, n_clusters):
    cluster_label = None
    if cluster_method == 'kmeans':
        km = KMeans(n_clusters=n_clusters, n_init=10,random_state=100, n_jobs=-1).fit(X)
        cluster_label = km.labels_
    elif cluster_method == 'em':
        em = GaussianMixture(n_components=n_clusters, covariance_type='diag', warm_start=True, random_state=100).fit(X)
        cluster_label = em.predict(X)
    df = pd.DataFrame(X)
    df['cluster'] = cluster_label
    label_cluster = LabelEncoder()
    df['cluster'] = label_cluster.fit_transform(df['cluster'])
    return np.array(df.values)
```

```python
newXpw_k = add_clusters(Xpw,'kmeans', 5)
newXpw_pca_k = add_clusters(pca_pw,'kmeans', 5)
newXpw_ica_k = add_clusters(ica_pw,'kmeans', 4)
newXpw_rp_k = add_clusters(rp_pw,'kmeans', 8)
newXpw_rfc_k = add_clusters(rfc_pw,'kmeans', 4)
```
